```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("/content/StudentsPerformance.csv")
```

```python
df.head()
```

```python
df["overall_average"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3
```

```python
gender_avg = df.groupby("gender")[
    "overall_average"
].mean()

print(gender_avg)
```

```python
lunch_avg = df.groupby("lunch")[
    "overall_average"
].mean()

print(lunch_avg)
```

```python
parent_avg = df.groupby(
    "parental level of education"
)["overall_average"].mean()

print(parent_avg)
```

```python
gender_avg.plot(kind="bar")

plt.ylabel("Average Score")

plt.show()
```

```python
lunch_avg.plot(kind="bar")

plt.ylabel("Average Score")

plt.show()
```

```python
parent_avg.plot(kind="bar")

plt.xticks(rotation=45)

plt.show()
```

```python
parent_avg.plot(kind="bar")

plt.xticks(rotation=45)

plt.show()
```

```python
group1 = df[df['gender'] == 'female']['overall_average']
group2 = df[df['gender'] == 'male']['overall_average']

t_stat, p_value = ttest_ind(group1, group2)

print("P-value:", p_value)
```

```python
p_value < 0.05
```
